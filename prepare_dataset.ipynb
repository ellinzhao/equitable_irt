## 2. Load all subject data

```python
dataset_dir = '/Users/ellin/Projects/therm/dataset'
name_dirs = os.listdir('/Users/ellin/Projects/therm/dataset/landmarks')
name_dirs = [name for name in name_dirs if name[0] != '.']
name_dirs = sorted(name_dirs)
```

```python
subjects = []
for name in tqdm(name_dirs[5:6]):
    print(name)
    sub = Subject(dataset_dir, name)
    subjects += [sub]
```

```python
plt.figure(figsize=(10, 10))
for i in range(len(subjects)):
    sub = subjects[i]
    j = np.arange(240).astype(int)
    j = j[~sub.base.invalid]
    j = j[20]
    plt.subplot(8, 9, i + 1)
    plt.imshow(sub.base.ir.warped[..., j], vmin=75)
    plt.axis('off')
    plt.title(sub.name)
```
